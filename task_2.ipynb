```python
data = data.rename(columns={
    "Marital Status":"Marital_Status",
    "Monthly Income":"Monthly_Income",
    "Educational Qualifications":"Educational_Qualifications",
    "Family size":"Family_size",
    "Pin_code":"Pin_code"
    })
```

```python
data.head()
```

```python
# ubah data categorikal ke numerik
data['Gender'] = data['Gender'].map({
    "Male":0,
    "Female":1
})
data['Marital_Status'] = data['Marital_Status'].map({
    "Single":0,
    "Married":1,
    "Prefer not to say":2
})
data['Occupation'] = data['Occupation'].map({
    "Student":0,
    "Employee":1,
    "Self Employeed":2,
    "House wife":3
})
data['Monthly_Income'] = data['Monthly_Income'].map({
    "No Income":0,
    "25001 to 50000":1,
    "More than 50000":2,
    "10001 to 25000":3,
    "Below Rs.10000":4
})
data['Educational_Qualifications'] = data['Educational_Qualifications'].map({
    "Graduate":0,
    "Post Graduate":1,
    "Ph.D":2,
    "School":3,
    "Uneducated":4
})
data['Output'] = data['Output'].map({
    "Yes":0,
    "No":1,
})
data['Feedback'] = data['Feedback'].map({
    "Positive":1,
    "Negative":0,
}).fillna(0)
```

```python
# Membagi dataset
atribut = ["Marital_Status","Age","Occupation","latitude","Monthly_Income"]
x = np.array(data[atribut])
y = np.array(data["Output"])
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
# membuat model
model = DecisionTreeClassifier()
model.fit(xtrain,ytrain)
```

```python
print(model.score(xtrain,ytrain))
print(model.score(xtest,ytest))
```
