```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import os
```

```python
data = pd.read_csv("D:\\belajar_data_science\\train_u6lujuX_CVtuZ9i (1).csv");
print(data.head())
```

```python
print(data.dtypes)
```

```python
# cek apakah ada missing value
print(data.isnull().sum())

# hapus missing value
data = data.dropna(how="any",axis=0)
```

```python
print(data.isnull().sum())
```

```python
data = data.drop(columns=["Loan_ID"])
```

```python
# ubah data categorikal ke numerik
data['Gender'] = data['Gender'].map({
    "Male":0,
    "Female":1
})
data['Married'] = data['Married'].map({"No":0,"Yes":1})
data['Education'] = data['Education'].map({"Not Graduate":0,"Graduate":1})
data['Self_Employed'] = data['Self_Employed'].map({"No":0,"Yes":1})
data['Property_Area'] = data['Property_Area'].map({"Semiurban":0,"Urban":1,"Rural":3})
data['Loan_Status'] = data['Loan_Status'].map({"N":0,"Y":1})
data['Dependents'] = data['Dependents'].map({"0":0,"1":1,"2":2,"3+":3})
```

```python
data.isnull().sum()
```

```python
print(data.head())
```

```python
correlation = data.corr()
correlation["Loan_Status"].sort_values(ascending=False)
```

```python
# Membagi dataset
atribut = ["Credit_History","Married","Education","Dependents"]
x = np.array(data[atribut])
y = np.array(data['Loan_Status'])
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
# Membuat model
model = DecisionTreeClassifier()
model.fit(xtrain,ytrain)
```

```python
print(model.score(xtrain,ytrain))
print(model.score(xtest,ytest))
```
